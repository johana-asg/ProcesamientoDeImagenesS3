```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft,fft2,ifft, fftshift
import matplotlib.pyplot as plt
import skimage
from skimage import io
import cv2 as cv
import math
```

```python
#Leer una imagen
Im = io.imread('miimagen.png')
plt.imshow(Im)
```

```python
#Extraemos un plano de la imagen (solo si la imagen es de 3 planos)
im = Im[:,:,2]

#Aplicamos TF y visualizamos
TF = np.fft.fftshift(np.fft.fft2(im))
plt.imshow(np.log(abs(TF)), cmap='gray')
```
